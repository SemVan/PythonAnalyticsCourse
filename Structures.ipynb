```python
# ========== СОЗДАНИЕ СПИСКОВ ==========
# Пустой список
empty_list = []
empty_list2 = list()

# Список с элементами
numbers = [1, 2, 3, 4, 5]
fruits = ["яблоко", "банан", "апельсин"]
mixed = [1, "текст", 3.14, True]

# Создание с помощью умножения
zeros = [0] * 5  # [0, 0, 0, 0, 0]

print("Создание списков:")
print(f"numbers: {numbers}")
print("numbers: {}".format(numbers))
print(f"fruits: {fruits}")
print(f"zeros: {zeros}")

# ========== ОСНОВНЫЕ ОПЕРАЦИИ ==========
print("\n=== Основные операции ===")

# Доступ к элементам (индексация)
print(f"Первый элемент: {fruits[0]}")      # яблоко
print(f"Последний элемент: {fruits[-1]}")  # апельсин

# Срезы (slicing)
print(f"Первые два: {fruits[0:2]}")        # ['яблоко', 'банан']
print(f"С шагом 2: {numbers[::3]}")        # [1, 3, 5]

# Изменение элементов
fruits[1] = "груша"
print(f"После изменения: {fruits}")

# ========== МЕТОДЫ СПИСКОВ ==========
print("\n=== Методы списков ===")

# Добавление элементов
fruits.append("виноград")           # в конец
fruits.insert(1, "киви") # по индексу
#[1, 2, 4]
print(f"После добавления: {fruits}")

# Удаление элементов
removed = fruits.pop()               # удаляет последний
removed_by_index = fruits.pop(0)     # удаляет по индексу
fruits.remove("груша")               # удаляет первое вхождение
print(f"После удаления: {fruits}")

# Поиск и информация
if "киви" in fruits:
    print("Киви есть в списке!")
print(f"Индекс киви: {fruits.index('киви')}")
print(f"Количество элементов: {len(fruits)}")

# Сортировка
numbers = [3, 1, 4, 1, 5, 9, 2]
numbers.sort()                       # сортировка на месте
print(f"Отсортированные числа: {numbers}")
numbers.sort(reverse=True)           # обратная сортировка
print(f"Обратная сортировка: {numbers}")

# Копирование списков (важная тонкость!)
original = [1, 2, 3]
shallow_copy = original              # ссылка на тот же объект!
real_copy = original.copy()          # настоящая копия
real_copy2 = original[:]             # тоже копия через срез

original[0] = 999
print(f"original: {original}")       # [999, 2, 3]
print(f"shallow_copy: {shallow_copy}") # [999, 2, 3] - изменился!
print(f"real_copy: {real_copy}")     # [1, 2, 3] - не изменился

# ========== ОПЕРАЦИИ СО СПИСКАМИ ==========
print("\n=== Операции со списками ===")

# Конкатенация
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = list1 + list2
print(f"Объединение: {combined}")

# Повторение
repeated = [1, 2] * 3
print(f"Повторение: {repeated}")

# Вложенные списки
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(f"Матрица: {matrix}")
print(f"Элемент матрицы: {matrix[1][2]}")  # 6
```

```python
# ========== СОЗДАНИЕ КОРТЕЖЕЙ ==========
print("\n" + "="*50)
print("КОРТЕЖИ")
print("="*50)

# Пустой кортеж
empty_tuple = ()
empty_tuple2 = tuple()

# Кортеж с одним элементом (важная тонкость!)
single_element = (42,)  # запятая обязательна!
not_a_tuple = (42)      # это просто число!

# Кортеж с несколькими элементами
coordinates = (10, 20)
person = ("Иван", "Иванов", 30)

print(f"coordinates: {coordinates}")
print(f"person: {person}")
print(f"single_element: {single_element} (тип: {type(single_element)})")

# ========== ОСНОВНЫЕ ОПЕРАЦИИ ==========
print("\n=== Основные операции с кортежами ===")

# Доступ к элементам
print(f"Имя: {person[0]}")
print(f"Фамилия: {person[1]}")
print(f"Возраст: {person[2]}")

# Срезы работают так же
print(f"Первые два элемента: {person[0:2]}")

# Распаковка кортежа (tuple unpacking)
name, surname, age = person
print(f"Распаковка: {name} {surname}, {age} лет")

# Обмен значений через кортежи
a, b = 10, 20
print(f"До обмена: a={a}, b={b}")
a, b = b, a  # красивый обмен значениями
print(f"После обмена: a={a}, b={b}")

# ========== НЕИЗМЕНЯЕМОСТЬ ==========
print("\n=== Неизменяемость кортежей ===")

# Это вызовет ошибку:
# coordinates[0] = 100  # TypeError!

# Но кортеж может содержать изменяемые объекты
mixed_tuple = ([1, 2, 3], "текст")
print(f"Смешанный кортеж: {mixed_tuple}")
mixed_tuple[0].append(4)  # можно изменить список внутри кортежа!
print(f"После изменения внутреннего списка: {mixed_tuple}")

# ========== МЕТОДЫ КОРТЕЖЕЙ ==========
print("\n=== Методы кортежей ===")

# У кортежей мало методов (только count и index)
repeated_numbers = (1, 2, 3, 2, 1, 2)
print(f"Количество двоек: {repeated_numbers.count(2)}")
print(f"Индекс первой тройки: {repeated_numbers.index(3)}")
```

```python
# ========== СОЗДАНИЕ СЛОВАРЕЙ ==========
print("\n" + "="*50)
print("СЛОВАРИ")
print("="*50)

# Пустой словарь
empty_dict = {}
empty_dict2 = dict()

# Словарь с элементами
student = {
    "имя": "Анна",
    "возраст": 20,
    "курс": 2,
    "предметы": ["математика", "физика"]
}

# Другой способ создания
colors = dict(красный="red", синий="blue", зеленый="green")

print(f"student: {student}")
print(f"colors: {colors}")

# ========== ОСНОВНЫЕ ОПЕРАЦИИ ==========
print("\n=== Основные операции со словарями ===")

# Доступ к элементам
print(f"Имя студента: {student['имя']}")
# print(student['фамилия'])  # KeyError!

# Безопасный доступ через get()
print(f"Фамилия: {student.get('фамилия', 'не указана')}")

# Проверка наличия ключа
if "возраст" in student:
    print("Возраст указан")

# ========== ИЗМЕНЕНИЕ СЛОВАРЕЙ ==========
print("\n=== Изменение словарей ===")

# Добавление/изменение элементов
student["стипендия"] = True
student["возраст"] = 21
print(f"После изменений: {student}")

# Удаление элементов
removed_value = student.pop("курс")  # удаляет и возвращает значение
del student["стипендия"]             # просто удаляет
print(f"После удаления: {student}")

# ========== МЕТОДЫ СЛОВАРЕЙ ==========
print("\n=== Методы словарей ===")

# Ключи, значения, пары
print(f"Ключи: {list(student.keys())}")
print(f"Значения: {list(student.values())}")
print(f"Пары: {list(student.items())}")

# Обновление словаря
updates = {"возраст": 22, "город": "Москва"}
student.update(updates)
print(f"После обновления: {student}")

# ========== ПЕРЕБОР СЛОВАРЕЙ ==========
print("\n=== Перебор словарей ===")

print("Ключи:")
for key in student:
    print(f"  {key}")

print("Пары ключ-значение:")
for key, value in student.items():
    print(f"  {key}: {value}")

# ========== ТОНКОСТИ СЛОВАРЕЙ ==========
print("\n=== Тонкости словарей ===")

# Ключами могут быть только хешируемые объекты
valid_dict = {
    "строка": "значение",
    123: "число",
    (1, 2): "кортеж"
    # [1, 2]: "список"  # TypeError - список не может быть ключом!
}

# Словарные включения (dict comprehension)
squares = {x: x*x for x in range(1, 6)}
print(f"Квадраты: {squares}")
```

```python
# ========== СОЗДАНИЕ МНОЖЕСТВ ==========
print("\n" + "="*50)
print("МНОЖЕСТВА")
print("="*50)

# Пустое множество (тонкость - {} создает словарь!)
empty_set = set()
# empty_set = {}  # это словарь, а не множество!

# Множество с элементами
numbers_set = {1, 2, 3, 4, 5}
fruits_set = {"яблоко", "банан", "апельсин"}

# Создание из списка (убирает дубликаты)
with_duplicates = [1, 2, 2, 3, 3, 3, 4]
unique_set = set(with_duplicates)
print(f"Было: {with_duplicates}")
print(f"Стало: {unique_set}")

# ========== ОСНОВНЫЕ ОПЕРАЦИИ ==========
print("\n=== Основные операции с множествами ===")

# Добавление элементов
fruits_set.add("киви")
fruits_set.add("банан")  # не добавится - уже есть
print(f"После добавления: {fruits_set}")

# Удаление элементов
fruits_set.remove("яблоко")  # KeyError если элемента нет
fruits_set.discard("груша")  # не вызывает ошибку если элемента нет
removed = fruits_set.pop()   # удаляет случайный элемент
print(f"После удаления: {fruits_set}")

# ========== ОПЕРАЦИИ НАД МНОЖЕСТВАМИ ==========
print("\n=== Операции над множествами ===")

set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

print(f"set_a: {set_a}")
print(f"set_b: {set_b}")

# Объединение
union = set_a | set_b  # или set_a.union(set_b)
print(f"Объединение: {union}")

# Пересечение
intersection = set_a & set_b  # или set_a.intersection(set_b)
print(f"Пересечение: {intersection}")

# Разность
difference = set_a - set_b  # или set_a.difference(set_b)
print(f"Разность A - B: {difference}")

# Симметрическая разность
symmetric_diff = set_a ^ set_b  # или set_a.symmetric_difference(set_b)
print(f"Симметрическая разность: {symmetric_diff}")

# ========== СРАВНЕНИЕ МНОЖЕСТВ ==========
print("\n=== Сравнение множеств ===")

small = {1, 2}
large = {1, 2, 3, 4}

print(f"small подмножество large: {small <= large}")  # True
print(f"large надмножество small: {large >= small}")  # True
print(f"Множества равны: {small == {1, 2}}")         # True

# ========== ТОНКОСТИ МНОЖЕСТВ ==========
print("\n=== Тонкости множеств ===")

# Множества могут содержать только хешируемые объекты
valid_set = {1, "текст", (1, 2)}
# invalid_set = {1, [1, 2]}  # TypeError - список не может быть элементом множества

# Множественные включения (set comprehension)
even_squares = {x*x for x in range(10) if x*x % 2 == 0}
print(f"Четные квадраты: {even_squares}")

# Замороженное множество (неизменяемое)
frozen = frozenset((1, 2, 3, 2, 1))
print(f"Замороженное множество: {frozen}")
# frozen.add(4)  # AttributeError - нельзя изменить!
d = {
    frozen: "experiment"
}
```
