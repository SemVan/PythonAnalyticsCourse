```python
#Задача 1: Проверка четности числа

# Get number from user
user_number = int(input("Введите целое число: "))

# Check if number is even or odd
if user_number % 6 == 0:
    print(f"Число {user_number} четное")
elif user_number % 3 == 0:
    print(f"Число {user_number} нечетное, но зато делится на 3")
else:
    print(f"Число {user_number} нечетное")

if user_number == 9:
    print('Кстати, это равно девятке')
elif user_number == 12:
    print('Кстати, это не равно девятке')
```

```python
#Задача 3: Поиск максимального из трех чисел

# Get three numbers from user
first_num = float(input("Введите первое число: "))
second_num = float(input("Введите второе число: "))
third_num = float(input("Введите третье число: "))

# Find maximum number
if first_num >= second_num and first_num >= third_num:
    max_number = first_num
elif second_num >= first_num and second_num >= third_num:
    max_number = second_num
else:
    max_number = third_num

print(f"Наибольшее число: {max_number}")
```

```python
#Задача 4: Таблица умножения


# Get number for multiplication table
base_number = int(input("Введите число для таблицы умножения: "))

print(f"Таблица умножения для {base_number}:")

# Generate multiplication table using for loop
for multiplier in range(1, 11):
    result = base_number * multiplier
    print(f"{base_number} × {multiplier} = {result}")
```

```python
#Задача 5: Проверка простого числа


# Get number from user
number_to_check = int(input("Введите число для проверки: "))

# Check if number is prime
is_prime = True

# Special cases
if number_to_check <= 1:
    is_prime = False
else:
    # Check divisors from 2 to number-1
    for divisor in range(2, number_to_check):
        if number_to_check % divisor == 0:
            is_prime = False
            break

# Display result
if is_prime:
    print(f"Число {number_to_check} является простым")
else:
    print(f"Число {number_to_check} не является простым")
```

```python
#Задача 6: Числа Фибоначчи


# Get how many Fibonacci numbers to display
fib_count = int(input("Сколько чисел Фибоначчи показать? "))

# Initialize first two Fibonacci numbers
first_fib = 0
second_fib = 1

print("Последовательность Фибоначчи:")

# Handle special cases
if fib_count >= 1:
    print(first_fib, end=" ")
if fib_count >= 2:
    print(second_fib, end=" ")

# Generate remaining Fibonacci numbers
for count in range(2, fib_count):
    next_fib = first_fib + second_fib
    print(next_fib, end=" ")
    
    # Update values for next iteration
    first_fib = second_fib
    second_fib = next_fib

print()  # New line at the end
```

```python
#Задача 7: Игра "Угадай число"

import random

# Generate random number between 1 and 100
secret_number = random.randint(1, 100)
attempts_count = 0
max_attempts = 7

print("Я загадал число от 1 до 100. У вас есть 7 попыток!")

# Main game loop
while attempts_count < max_attempts:
    # Get user's guess
    user_guess = int(input("Ваша догадка: "))
    attempts_count += 1
    
    # Check the guess
    if user_guess < secret_number:
        print("Слишком маленькое число!")
    elif user_guess > secret_number:
        print("Слишком большое число!")
    else:
        print(f"Поздравляю! Вы угадали число {secret_number} за {attempts_count} попыток!")
        break
    
    # Show remaining attempts
    remaining_attempts = max_attempts - attempts_count
    if remaining_attempts > 0 and user_guess != secret_number:
        print(f"Осталось попыток: {remaining_attempts}")

# Game over message
if attempts_count == max_attempts and user_guess != secret_number:
    print(f"Игра окончена! Загаданное число было: {secret_number}")
```

```python
#Задача 8: Анализ последовательности чисел


print("Вводите числа (0 для завершения):")

# Initialize counters
positive_count = 0
negative_count = 0
zero_count = 0
total_sum = 0
number_count = 0

# Process numbers until 0 is entered
while True:
    current_number = float(input("Число: "))
    
    # Check for termination condition
    if current_number == 0:
        break
    
    # Update counters based on number type
    if current_number > 0:
        positive_count += 1
    elif current_number < 0:
        negative_count += 1
    else:
        zero_count += 1
    
    # Update statistics
    total_sum += current_number
    number_count += 1

# Calculate average
if number_count > 0:
    average_value = total_sum / number_count
else:
    average_value = 0

# Display results
print("\nРезультаты анализа:")
print(f"Всего чисел: {number_count}")
print(f"Положительных: {positive_count}")
print(f"Отрицательных: {negative_count}")
print(f"Нулевых: {zero_count}")
print(f"Среднее значение: {average_value:.2f}")
```
