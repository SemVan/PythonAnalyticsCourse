```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Настройка для отображения графиков в ноутбуке
%matplotlib inline

def basic_plots():
    """Самые базовые графики"""
    
    # Данные для примеров
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    
    print("=== БАЗОВЫЕ ГРАФИКИ MATPLOTLIB ===")
    
    # 1. Линейный график
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='sin(x)')
    plt.title('Простой линейный график')
    plt.xlabel('X ось')
    plt.ylabel('Y ось')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # 2. Точечный график
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, alpha=0.6, color='red')
    plt.title('Точечный график')
    plt.xlabel('X ось')
    plt.ylabel('Y ось')
    plt.show()
    
    # 3. Столбчатая диаграмма
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    
    plt.figure(figsize=(10, 6))
    plt.bar(categories, values, color=['blue', 'green', 'red', 'orange'])
    plt.title('Столбчатая диаграмма')
    plt.xlabel('Категории')
    plt.ylabel('Значения')
    plt.show()
    
    # 4. Гистограмма
    data = np.random.normal(0, 1, 1000)
    
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=30, alpha=0.7, color='purple', edgecolor='black')
    plt.title('Гистограмма нормального распределения')
    plt.xlabel('Значения')
    plt.ylabel('Частота')
    plt.show()

basic_plots()
```

```python
def multiple_plots():
    """Создание нескольких графиков на одной фигуре"""
    
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    y3 = np.tan(x)
    y4 = np.exp(-x)
    
    print("=== МНОЖЕСТВЕННЫЕ ГРАФИКИ ===")
    
    # 1. Несколько линий на одном графике
    plt.figure(figsize=(12, 8))
    
    plt.subplot(2, 2, 1)  # 2 строки, 2 столбца, 1-й график
    plt.plot(x, y1, 'b-', label='sin(x)')
    plt.plot(x, y2, 'r--', label='cos(x)')
    plt.title('sin и cos')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(2, 2, 2)
    plt.plot(x, y3, 'g-', label='tan(x)')
    plt.title('tan(x)')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(2, 2, 3)
    plt.plot(x, y4, 'm-', label='exp(-x)')
    plt.title('Экспонента')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(2, 2, 4)
    plt.plot(x, y1, 'b-', label='sin(x)')
    plt.plot(x, y2, 'r--', label='cos(x)')
    plt.plot(x, y3, 'g:', label='tan(x)')
    plt.plot(x, y4, 'm-.', label='exp(-x)')
    plt.title('Все функции')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # 2. Использование subplots
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    
    axes[0, 0].plot(x, y1, color='blue')
    axes[0, 0].set_title('Синус')
    
    axes[0, 1].scatter(x, y2, color='red', alpha=0.5)
    axes[0, 1].set_title('Косинус (точки)')
    
    axes[1, 0].bar(range(len(y3[:10])), y3[:10], color='green')
    axes[1, 0].set_title('Тангенс (первые 10)')
    
    axes[1, 1].hist(y4, bins=20, color='orange', alpha=0.7)
    axes[1, 1].set_title('Экспонента (гистограмма)')
    
    plt.tight_layout()
    plt.show()

multiple_plots()
```

```python
def styling_and_customization():
    """Кастомизация графиков"""
    
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    
    print("=== КАСТОМИЗАЦИЯ ГРАФИКОВ ===")
    
    # 1. Разные стили линий и маркеров
    plt.figure(figsize=(12, 8))
    
    plt.subplot(2, 2, 1)
    plt.plot(x, y, linewidth=3, linestyle='-', color='blue', label='solid')
    plt.legend()
    
    plt.subplot(2, 2, 2)
    plt.plot(x, y, linewidth=2, linestyle='--', color='red', label='dashed')
    plt.legend()
    
    plt.subplot(2, 2, 3)
    plt.plot(x, y, linewidth=2, linestyle=':', color='green', label='dotted')
    plt.legend()
    
    plt.subplot(2, 2, 4)
    plt.plot(x, y, marker='o', markersize=3, linestyle='-', color='purple', label='with markers')
    plt.legend()
    
    plt.suptitle('Разные стили линий')
    plt.tight_layout()
    plt.show()
    
    # 2. Настройка осей и сетки
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, linewidth=2)
    
    # Настройка осей
    plt.xlim(2, 8)
    plt.ylim(-1.5, 1.5)
    plt.xticks(np.arange(2, 9, 1))
    plt.yticks(np.arange(-1, 1.5, 0.5))
    
    # Сетка
    plt.grid(True, alpha=0.3, linestyle='--')
    
    # Подписи
    plt.title('Кастомизированный график', fontsize=14, fontweight='bold')
    plt.xlabel('X значения', fontsize=12)
    plt.ylabel('Y значения', fontsize=12)
    
    plt.show()
    
    # 3. Разные маркеры
    markers = ['o', 's', '^', 'D', 'v', '<', '>', 'p', '*', 'h']
    
    plt.figure(figsize=(12, 6))
    for i, marker in enumerate(markers):
        plt.plot(i, 0, marker=marker, markersize=10, label=marker, linestyle='')
    
    plt.xlim(-1, len(markers))
    plt.legend()
    plt.title('Разные типы маркеров')
    plt.show()

styling_and_customization()
```

```python
def introduction_to_seaborn():
    """Базовые графики Seaborn"""
    
    # Загрузка встроенных датасетов
    tips = sns.load_dataset('tips')
    iris = sns.load_dataset('iris')
    
    print("=== БАЗОВЫЕ ГРАФИКИ SEABORN ===")
    print("Данные tips:")
    print(tips.head())
    
    # 1. Distribution plot
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 3, 1)
    sns.histplot(data=tips, x='total_bill', kde=True)
    plt.title('Гистограмма с KDE')
    
    plt.subplot(1, 3, 2)
    sns.kdeplot(data=tips, x='total_bill', hue='time', fill=True)
    plt.title('KDE по времени')
    
    plt.subplot(1, 3, 3)
    sns.boxplot(data=tips, x='time', y='total_bill')
    plt.title('Boxplot по времени')
    
    plt.tight_layout()
    plt.show()
    
    # 2. Scatter plot с дополнительными измерениями
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=tips, x='total_bill', y='tip', 
                   hue='time', style='smoker', size='size')
    plt.title('Типы vs Счет: время, курение и размер компании')
    plt.show()
    
    # 3. Pairplot для анализа взаимосвязей
    print("\nPairplot для iris dataset:")
    sns.pairplot(iris, hue='species')
    plt.show()

introduction_to_seaborn()
```

```python
def statistical_plots():
    """Статистические визуализации в Seaborn"""
    
    tips = sns.load_dataset('tips')
    titanic = sns.load_dataset('titanic')
    
    print("=== СТАТИСТИЧЕСКИЕ ГРАФИКИ ===")
    
    # 1. Bar plot с доверительными интервалами
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    sns.barplot(data=tips, x='day', y='total_bill', ci='sd')
    plt.title('Средний счет по дням с SD')
    
    plt.subplot(1, 2, 2)
    sns.barplot(data=tips, x='day', y='total_bill', hue='sex')
    plt.title('Счет по дням и полу')
    
    plt.tight_layout()
    plt.show()
    
    # 2. Violin plot
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    sns.violinplot(data=tips, x='day', y='total_bill')
    plt.title('Violin plot по дням')
    
    plt.subplot(1, 2, 2)
    sns.violinplot(data=tips, x='day', y='total_bill', hue='smoker', split=True)
    plt.title('Violin plot с разделением по курению')
    
    plt.tight_layout()
    plt.show()
    
    # 3. Heatmap корреляций
    plt.figure(figsize=(10, 8))
    
    # Вычисляем корреляционную матрицу
    numeric_tips = tips.select_dtypes(include=[np.number])
    correlation_matrix = numeric_tips.corr()
    
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
                square=True, linewidths=0.5)
    plt.title('Матрица корреляций')
    plt.show()
    
    # 4. Count plot
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    sns.countplot(data=titanic, x='class', hue='alive')
    plt.title('Выживаемость по классам (Titanic)')
    
    plt.subplot(1, 2, 2)
    sns.countplot(data=titanic, x='deck', hue='alive', order=sorted(titanic['deck'].dropna().unique()))
    plt.title('Выживаемость по палубам')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()

statistical_plots()
```

```python
def advanced_plots():
    """Продвинутые типы графиков"""
    
    # Создаем сложные данные
    np.random.seed(42)
    x = np.random.randn(1000)
    y = np.random.randn(1000)
    categories = np.random.choice(['A', 'B', 'C'], 1000)
    values = np.random.rand(1000) * 100
    
    print("=== ПРОДВИНУТЫЕ ГРАФИКИ ===")
    
    # 1. 2D Histogram (Hexbin)
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.hexbin(x, y, gridsize=30, cmap='Blues')
    plt.colorbar()
    plt.title('2D Histogram (Hexbin)')
    plt.xlabel('X')
    plt.ylabel('Y')
    
    plt.subplot(1, 2, 2)
    sns.kdeplot(x=x, y=y, fill=True, cmap='Reds', thresh=0.1)
    plt.title('2D KDE Plot')
    
    plt.tight_layout()
    plt.show()
    
    # 2. Stacked area chart
    time = np.arange(50)
    data1 = np.cumsum(np.random.randn(50)) + 20
    data2 = np.cumsum(np.random.randn(50)) + 10
    data3 = np.cumsum(np.random.randn(50)) + 5
    
    plt.figure(figsize=(10, 6))
    plt.stackplot(time, data1, data2, data3, 
                  labels=['Series 1', 'Series 2', 'Series 3'],
                  alpha=0.7)
    plt.legend(loc='upper left')
    plt.title('Stacked Area Chart')
    plt.show()
    
    # 3. Polar plot
    theta = np.linspace(0, 2 * np.pi, 8, endpoint=False)
    radii = np.random.rand(8) * 10
    
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, projection='polar')
    ax.bar(theta, radii, alpha=0.7, color=np.random.rand(8, 3))
    ax.set_title('Polar Bar Chart')
    plt.show()

advanced_plots()
```

```python
def time_series_plots():
    """Визуализация временных рядов"""
    
    # Создаем временной ряд
    dates = pd.date_range('2023-01-01', periods=100, freq='D')
    values = np.cumsum(np.random.randn(100)) + 100
    volume = np.random.randint(100, 1000, 100)
    
    df = pd.DataFrame({
        'date': dates,
        'price': values,
        'volume': volume
    })
    df.set_index('date', inplace=True)
    
    print("=== ВРЕМЕННЫЕ РЯДЫ ===")
    
    # 1. Простой временной ряд
    plt.figure(figsize=(12, 8))
    
    plt.subplot(2, 2, 1)
    plt.plot(df.index, df['price'], linewidth=2)
    plt.title('Цена во времени')
    plt.xticks(rotation=45)
    
    # 2. С скользящим средним
    plt.subplot(2, 2, 2)
    df['MA_7'] = df['price'].rolling(window=7).mean()
    df['MA_30'] = df['price'].rolling(window=30).mean()
    
    plt.plot(df.index, df['price'], alpha=0.5, label='Цена')
    plt.plot(df.index, df['MA_7'], label='7-дневное среднее')
    plt.plot(df.index, df['MA_30'], label='30-дневное среднее')
    plt.title('Цена со скользящими средними')
    plt.legend()
    plt.xticks(rotation=45)
    
    # 3. Область с доверительным интервалом
    plt.subplot(2, 2, 3)
    
    # Создаем искусственный доверительный интервал
    upper = df['price'] + np.random.randn(100) * 2
    lower = df['price'] - np.random.randn(100) * 2
    
    plt.plot(df.index, df['price'], color='black', linewidth=1)
    plt.fill_between(df.index, lower, upper, alpha=0.3, color='blue')
    plt.title('Цена с доверительным интервалом')
    plt.xticks(rotation=45)
    
    # 4. Сезонность
    plt.subplot(2, 2, 4)
    seasonal = np.sin(2 * np.pi * np.arange(100) / 30) * 5 + df['price']
    plt.plot(df.index, seasonal)
    plt.title('Временной ряд с сезонностью')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()

time_series_plots()
```

```python
def customization_and_themes():
    """Кастомизация и темы оформления"""
    
    # Данные для примеров
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    
    print("=== ТЕМЫ И КАСТОМИЗАЦИЯ ===")
    
    # 1. Разные стили matplotlib
    styles = ['default', 'classic', 'ggplot', 'fivethirtyeight']
    
    for style in styles:
        plt.style.use(style)
        plt.figure(figsize=(8, 4))
        plt.plot(x, y1, label='sin(x)')
        plt.plot(x, y2, label='cos(x)')
        plt.title(f'Стиль: {style}')
        plt.legend()
        plt.show()
    
    # Вернем стиль по умолчанию
    plt.style.use('default')
    
    # 2. Seaborn темы
    sns_themes = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
    
    for theme in sns_themes:
        sns.set_style(theme)
        plt.figure(figsize=(8, 4))
        sns.lineplot(x=x, y=y1, label='sin(x)')
        sns.lineplot(x=x, y=y2, label='cos(x)')
        plt.title(f'Seaborn тема: {theme}')
        plt.legend()
        plt.show()
    
    # 3. Цветовые палитры
    palettes = ['viridis', 'plasma', 'inferno', 'magma', 'coolwarm']
    
    plt.figure(figsize=(15, 10))
    for i, palette in enumerate(palettes, 1):
        plt.subplot(2, 3, i)
        data = np.random.rand(10, 10)
        sns.heatmap(data, cmap=palette, cbar=True)
        plt.title(f'Палитра: {palette}')
    
    plt.tight_layout()
    plt.show()
    
    # 4. Кастомизация фигуры
    fig = plt.figure(figsize=(12, 8), facecolor='lightgray')
    ax = fig.add_subplot(111, facecolor='white')
    
    # Создаем сложный график
    x = np.random.randn(1000)
    y = np.random.randn(1000)
    colors = np.random.rand(1000)
    sizes = np.random.rand(1000) * 100
    
    scatter = ax.scatter(x, y, c=colors, s=sizes, alpha=0.6, 
                        cmap='viridis', edgecolors='black', linewidth=0.5)
    
    # Цветовая шкала
    cbar = fig.colorbar(scatter, ax=ax)
    cbar.set_label('Цветовые значения')
    
    # Настройка осей
    ax.set_xlabel('X ось', fontsize=12, fontweight='bold')
    ax.set_ylabel('Y ось', fontsize=12, fontweight='bold')
    ax.set_title('Кастомизированный scatter plot', fontsize=14, fontweight='bold')
    
    # Сетка
    ax.grid(True, alpha=0.3, linestyle='--')
    
    # Легенда для размеров
    for size in [20, 50, 80]:
        ax.scatter([], [], c='gray', alpha=0.6, s=size, 
                  label=str(size), edgecolors='black')
    ax.legend(scatterpoints=1, frameon=True, labelspacing=1, title='Размеры')
    
    plt.show()

customization_and_themes()
```

```python
def three_d_plots():
    """Построение поверхностей"""
    
    try:
        from matplotlib.animation import FuncAnimation
        from mpl_toolkits.mplot3d import Axes3D
        
        print("=== ПОСТРОЕНИЕ ПОВЕРХНОСТЕЙ ===")
        
        # 1. 3D график
        fig = plt.figure(figsize=(10, 8))
        ax = fig.add_subplot(111, projection='3d')
        
        # Создаем данные для 3D поверхности
        x = np.linspace(-5, 5, 50)
        y = np.linspace(-5, 5, 50)
        X, Y = np.meshgrid(x, y)
        Z = np.sin(np.sqrt(X**2 + Y**2))
        
        surface = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')
        ax.set_title('3D Surface Plot')
        fig.colorbar(surface)
        plt.show()
        
    except ImportError as e:
        print(f"Некоторые функции требуют дополнительных зависимостей: {e}")

three_d_plots()
```

```python
def complete_data_analysis():
    """Полный пример анализа данных с визуализацией"""
    
    # Создаем синтетический датасет
    np.random.seed(42)
    n = 500
    
    data = pd.DataFrame({
        'age': np.random.normal(35, 10, n).astype(int),
        'income': np.random.normal(50000, 15000, n),
        'spending_score': np.random.randint(1, 100, n),
        'gender': np.random.choice(['Male', 'Female'], n),
        'city': np.random.choice(['NY', 'LA', 'Chicago', 'Miami'], n)
    })
    
    # Ограничим доход положительными значениями
    data['income'] = data['income'].clip(lower=10000)
    
    print("=== ПОЛНЫЙ АНАЛИЗ ДАННЫХ ===")
    print("Первые 5 строк данных:")
    print(data.head())
    
    # Создаем комплексную фигуру
    fig = plt.figure(figsize=(20, 16))
    
    # 1. Распределение возраста
    ax1 = plt.subplot(3, 3, 1)
    sns.histplot(data=data, x='age', kde=True, ax=ax1)
    ax1.set_title('Распределение возраста')
    
    # 2. Доход по городам
    ax2 = plt.subplot(3, 3, 2)
    sns.boxplot(data=data, x='city', y='income', ax=ax2)
    ax2.set_title('Доход по городам')
    ax2.tick_params(axis='x', rotation=45)
    
    # 3. Spending score по полу
    ax3 = plt.subplot(3, 3, 3)
    sns.violinplot(data=data, x='gender', y='spending_score', ax=ax3)
    ax3.set_title('Spending Score по полу')
    
    # 4. Возраст vs Доход
    ax4 = plt.subplot(3, 3, 4)
    scatter = sns.scatterplot(data=data, x='age', y='income', 
                            hue='city', size='spending_score', ax=ax4)
    ax4.set_title('Возраст vs Доход')
    
    # 5. Корреляционная матрица
    ax5 = plt.subplot(3, 3, 5)
    numeric_data = data.select_dtypes(include=[np.number])
    corr_matrix = numeric_data.corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, ax=ax5)
    ax5.set_title('Матрица корреляций')
    
    # 6. Совместное распределение возраста и дохода
    ax6 = plt.subplot(3, 3, 6)
    sns.kdeplot(data=data, x='age', y='income', fill=True, cmap='Blues', ax=ax6)
    ax6.set_title('Совместное распределение: Возраст-Доход')
    
    # 7. Средние значения по городам
    ax7 = plt.subplot(3, 3, 7)
    city_stats = data.groupby('city').agg({
        'income': 'mean',
        'spending_score': 'mean',
        'age': 'mean'
    }).reset_index()
    
    x = np.arange(len(city_stats))
    width = 0.25
    
    ax7.bar(x - width, city_stats['income'], width, label='Доход', alpha=0.7)
    ax7.bar(x, city_stats['spending_score'], width, label='Spending Score', alpha=0.7)
    ax7.bar(x + width, city_stats['age'], width, label='Возраст', alpha=0.7)
    
    ax7.set_xticks(x)
    ax7.set_xticklabels(city_stats['city'])
    ax7.legend()
    ax7.set_title('Средние показатели по городам')
    
    # 8. Количество людей по городам и полу
    ax8 = plt.subplot(3, 3, 8)
    count_data = data.groupby(['city', 'gender']).size().unstack()
    count_data.plot(kind='bar', ax=ax8)
    ax8.set_title('Распределение по городам и полу')
    ax8.tick_params(axis='x', rotation=45)
    
    # 9. Cumulative distribution function
    ax9 = plt.subplot(3, 3, 9)
    sorted_income = np.sort(data['income'])
    y_vals = np.arange(1, len(sorted_income) + 1) / len(sorted_income)
    ax9.plot(sorted_income, y_vals, linewidth=2)
    ax9.set_xlabel('Доход')
    ax9.set_ylabel('CDF')
    ax9.set_title('CDF дохода')
    ax9.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Дополнительный анализ
    print("\n=== СВОДНАЯ СТАТИСТИКА ===")
    print(data.describe())
    
    print("\n=== КОРРЕЛЯЦИИ ===")
    print(corr_matrix)

complete_data_analysis()
```
