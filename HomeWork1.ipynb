```python
def analyze_numbers(numbers):
    if not numbers:  # Проверка на пустой список
        return {
            'sum': 0,
            'count': 0,
            'avg': 0,
            'max': 0,
            'min': 0,
            'even_count': 0,
            'positive_count': 0
        }
    
    total_sum = sum(numbers)
    count = len(numbers)
    avg = total_sum / count
    max_val = max(numbers)
    min_val = min(numbers)
    
    even_count = 0
    positive_count = 0
    for num in numbers:
        if num % 2 == 0:
            even_count += 1
        if num > 0:
            positive_count += 1
    
    return {
        'sum': total_sum,
        'count': count,
        'avg': avg,
        'max': max_val,
        'min': min_val,
        'even_count': even_count,
        'positive_count': positive_count
    }

# Пример вызова
data = [1, -2, 3, 4, -5, 6]
result = analyze_numbers(data)
print(result)
```

```python
def convert_temperature(value, unit_from, unit_to):
    # Проверяем корректность единиц
    if unit_from not in ['C', 'F', 'K'] or unit_to not in ['C', 'F', 'K']:
        return None
    
    # Приводим к Цельсию как промежуточной единице
    if unit_from == 'C':
        temp_c = value
    elif unit_from == 'F':
        temp_c = (value - 32) * 5 / 9
    else:  # K
        temp_c = value - 273.15
    
    # Переводим из Цельсия в нужную единицу
    if unit_to == 'C':
        return temp_c
    elif unit_to == 'F':
        return temp_c * 9 / 5 + 32
    else:  # K
        return temp_c + 273.15

# Примеры вызовов
print(convert_temperature(100, 'C', 'F'))  # 212.0
print(convert_temperature(32, 'F', 'C'))   # 0.0
print(convert_temperature(0, 'C', 'K'))    # 273.15
```

```python
def check_password(password):
    if len(password) < 8:
        return False
    if ' ' in password:
        return False
    
    has_digit = False
    has_upper = False
    has_lower = False
    
    for char in password:
        if char.isdigit():
            has_digit = True
        elif char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
    
    return has_digit and has_upper and has_lower

# Примеры вызовов
print(check_password("Password123"))  # True
print(check_password("password"))       # False (нет цифры и заглавной)
print(check_password("Pass123"))        # False (меньше 8 символов)
```

```python
import string

def count_words(text):
    # Очищаем текст от знаков препинания
    cleaned_text = ""
    for char in text:
        if char in string.ascii_letters or char == ' ':
            cleaned_text += char.lower()
    
    # Разбиваем на слова и фильтруем пустые строки
    words = [word for word in cleaned_text.split() if word]
    
    # Считаем вхождения
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
    
    return word_count

# Пример вызова
text = "Hello, world# Hello Python. Python is great!"
result = count_words(text)
print(result)  # {'hello': 2, 'world': 1, 'python': 2, 'is': 1, 'great': 1}
```

```python
def print_multiplication_table(n):
    for i in range(1, n + 1):
        row = ""
        for j in range(1, n + 1):
            product = i * j
            row += f"{product:4}"  # 4 символа, выравнивание вправо
        print(row)

# Пример вызова
print_multiplication_table(5)
```
