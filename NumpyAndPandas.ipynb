```python
import numpy as np
import os

# Создаем директорию
if not os.path.exists('numpy_data'):
    os.makedirs('numpy_data')

# Генерация финансовых данных чисто на NumPy
np.random.seed(42)

# Параметры
n_records = 100
companies = ['AAPL', 'GOOGL', 'MSFT', 'TSLA', 'AMZN']
n_companies = len(companies)

# Создаем структурированный массив NumPy
dtype = [
    ('date', 'U10'),
    ('company', 'U10'),
    ('open_price', 'f8'),
    ('close_price', 'f8'),
    ('volume', 'i8'),
    ('pe_ratio', 'f8'),
    ('dividend_yield', 'f8')
]

# Генерация данных
financial_data = np.zeros(n_records, dtype=dtype)

for i in range(n_records):
    company = companies[i % n_companies]
    base_price = np.random.uniform(50, 500)
    
    financial_data[i] = (
        f'2024-01-{(i % 30) + 1:02d}',  # date
        company,                        # company
        round(base_price, 2),           # open_price
        round(base_price * np.random.uniform(0.95, 1.05), 2),  # close_price
        np.random.randint(1000000, 50000000),  # volume
        round(np.random.uniform(10, 40), 2),   # pe_ratio
        round(np.random.uniform(0, 0.05), 4)   # dividend_yield
    )

# Сохраняем в CSV чисто средствами NumPy
header = 'date,company,open_price,close_price,volume,pe_ratio,dividend_yield'
np.savetxt('numpy_data/stock_data_numpy.csv', financial_data, 
           fmt='%s,%s,%.2f,%.2f,%d,%.2f,%.4f',
           header=header, comments='')

print("Данные сгенерированы и сохранены чисто на NumPy!")
print("Первые 5 записей:")
for i in range(5):
    print(financial_data[i])
```

```python
import numpy as np

print("=== ЧИСТЫЙ NUMPY АНАЛИЗ ===\n")

# Загрузка данных чисто NumPy
data = np.genfromtxt('numpy_data/stock_data_numpy.csv', 
                     delimiter=',', dtype=None, names=True, encoding='utf-8')

print("1. ОБЗОР ДАННЫХ:")
print(f"Размер данных: {data.shape}")
print(f"Поля: {data.dtype.names}")
print(f"\nПервые 3 записи:")
for i in range(3):
    print(data[i])

# 2. БАЗОВЫЕ СТАТИСТИКИ
print("\n" + "="*50)
print("2. БАЗОВЫЕ СТАТИСТИКИ")

close_prices = data['close_price']
open_prices = data['open_price']
volumes = data['volume']
pe_ratios = data['pe_ratio']
dividends = data['dividend_yield']

print(f"Средняя цена закрытия: {np.mean(close_prices):.2f}")
print(f"Медианная цена: {np.median(close_prices):.2f}")
print(f"Стандартное отклонение цен: {np.std(close_prices):.2f}")
print(f"Минимальная цена: {np.min(close_prices):.2f}")
print(f"Максимальная цена: {np.max(close_prices):.2f}")

# 3. РАСЧЕТ ДОХОДНОСТИ
print("\n" + "="*50)
print("3. РАСЧЕТ ДОХОДНОСТИ")

daily_returns = ((close_prices - open_prices) / open_prices) * 100
print(f"Средняя дневная доходность: {np.mean(daily_returns):.2f}%")
print(f"Волатильность: {np.std(daily_returns):.2f}%")

# 4. ФИЛЬТРАЦИЯ
print("\n" + "="*50)
print("4. ФИЛЬТРАЦИЯ ДАННЫХ")

# Акции с доходностью > 2%
high_return_mask = daily_returns > 2.0
high_return_data = data[high_return_mask]
print(f"Акции с доходностью > 2%: {len(high_return_data)}")

# Недооцененные акции (P/E < 15)
undervalued_mask = pe_ratios < 15
undervalued_data = data[undervalued_mask]
print(f"Недооцененные акции (P/E < 15): {len(undervalued_data)}")

# Акции с дивидендами
dividend_mask = dividends > 0.02
dividend_data = data[dividend_mask]
print(f"Акции с дивидендами > 2%: {len(dividend_data)}")

# 5. СТАТИСТИЧЕСКИЙ АНАЛИЗ
print("\n" + "="*50)
print("5. СТАТИСТИЧЕСКИЙ АНАЛИЗ")

# Z-score для выявления аномалий
z_scores = (daily_returns - np.mean(daily_returns)) / np.std(daily_returns)
anomalies_mask = np.abs(z_scores) > 2
anomalies = data[anomalies_mask]
print(f"Аномальные движения (|Z-score| > 2): {len(anomalies)}")

# Value at Risk (95%)
var_95 = np.percentile(daily_returns, 5)
print(f"Value at Risk (95%): {var_95:.2f}%")

# 6. ГРУППИРОВКА ПО КОМПАНИЯМ (на NumPy)
print("\n" + "="*50)
print("6. АНАЛИЗ ПО КОМПАНИЯМ")

companies = np.unique(data['company'])
print("\nСтатистика по компаниям:")
print("Компания | Ср.цена | Ср.доходность | Ср.P/E")

for company in companies:
    company_mask = data['company'] == company
    company_data = data[company_mask]
    
    avg_price = np.mean(company_data['close_price'])
    avg_return = np.mean(daily_returns[company_mask])
    avg_pe = np.mean(company_data['pe_ratio'])
    
    print(f"{company:8} | {avg_price:7.2f} | {avg_return:13.2f}% | {avg_pe:6.2f}")

# 7. КОНКАТЕНАЦИЯ ДАННЫХ
print("\n" + "="*50)
print("7. КОНКАТЕНАЦИЯ И СОХРАНЕНИЕ")

# Создаем новый массив с добавленными полями
new_dtype = data.dtype.descr + [('daily_return', 'f8'), ('z_score', 'f8'), ('is_anomaly', 'bool')]
enhanced_data = np.zeros(data.shape, dtype=new_dtype)

# Копируем существующие данные
for field in data.dtype.names:
    enhanced_data[field] = data[field]

# Добавляем новые поля
enhanced_data['daily_return'] = daily_returns
enhanced_data['z_score'] = z_scores
enhanced_data['is_anomaly'] = anomalies_mask

# Сохраняем обогащенные данные
header_enhanced = 'date,company,open_price,close_price,volume,pe_ratio,dividend_yield,daily_return,z_score,is_anomaly'
np.savetxt('numpy_data/enhanced_data_numpy.csv', enhanced_data, 
           fmt='%s,%s,%.2f,%.2f,%d,%.2f,%.4f,%.2f,%.2f,%s',
           header=header_enhanced, comments='')

print("Обогащенные данные сохранены!")

# 8. КОМПЛЕКСНАЯ ФИЛЬТРАЦИЯ
print("\n" + "="*50)
print("8. КОМПЛЕКСНАЯ ФИЛЬТРАЦИЯ")

# Идеальные акции: недооцененные + положительная доходность + не аномальные
ideal_mask = (
    (pe_ratios < 20) & 
    (daily_returns > 0) & 
    (np.abs(z_scores) < 1.5) &
    (dividends > 0.01)
)

ideal_stocks = data[ideal_mask]
print(f"Найдено идеальных акций: {len(ideal_stocks)}")

if len(ideal_stocks) > 0:
    print("\nРекомендуемые акции:")
    for stock in ideal_stocks:
        return_pct = ((stock['close_price'] - stock['open_price']) / stock['open_price']) * 100
        print(f"  {stock['company']}: цена {stock['close_price']:.2f}, доходность {return_pct:.2f}%, P/E {stock['pe_ratio']:.2f}")

print(f"\nАнализ завершен! Обработано {len(data)} записей.")
```

```python
import pandas as pd
import numpy as np
import os

print("=== ЧИСТЫЙ PANDAS АНАЛИЗ ===\n")

# Создаем тестовые данные чисто Pandas
if not os.path.exists('pandas_data'):
    os.makedirs('pandas_data')

# Генерация данных
np.random.seed(123)
n_records = 100

data = {
    'date': pd.date_range('2024-01-01', periods=n_records),
    'company': np.random.choice(['AAPL', 'GOOGL', 'MSFT', 'TSLA', 'AMZN'], n_records),
    'sector': np.random.choice(['Tech', 'Auto', 'Retail'], n_records),
    'open_price': np.round(np.random.uniform(50, 500, n_records), 2),
    'close_price': np.round(np.random.uniform(45, 550, n_records), 2),
    'volume': np.random.randint(1000000, 50000000, n_records),
    'pe_ratio': np.round(np.random.uniform(10, 40, n_records), 2),
    'dividend_yield': np.round(np.random.uniform(0, 0.05, n_records), 4)
}

df = pd.DataFrame(data)

# Добавляем немного логики в цены
df['close_price'] = df['open_price'] * np.random.uniform(0.9, 1.1, n_records)

# Сохраняем исходные данные
df.to_csv('pandas_data/stock_data_pandas.csv', index=False)

print("1. ЗАГРУЗКА И ОБЗОР ДАННЫХ:")
df = pd.read_csv('pandas_data/stock_data_pandas.csv')
print(f"Размер данных: {df.shape}")
print(f"\nИнформация о данных:")
print(df.info())
print(f"\nПервые 5 записей:")
print(df.head())

# 2. БАЗОВЫЙ АНАЛИЗ PANDAS
print("\n" + "="*50)
print("2. БАЗОВЫЙ АНАЛИЗ")

print("Статистика по числовым колонкам:")
print(df.describe())

print(f"\nКоличество записей по компаниям:")
print(df['company'].value_counts())

# 3. ДОБАВЛЕНИЕ ВЫЧИСЛЯЕМЫХ ПОЛЕЙ
print("\n" + "="*50)
print("3. ВЫЧИСЛЯЕМЫЕ ПОЛЯ")

df['daily_return'] = ((df['close_price'] - df['open_price']) / df['open_price']) * 100
df['price_range'] = df['close_price'] - df['open_price']
df['is_profitable'] = df['daily_return'] > 0

print("Данные с добавленными полями:")
print(df[['company', 'open_price', 'close_price', 'daily_return', 'is_profitable']].head())

# 4. ФИЛЬТРАЦИЯ PANDAS
print("\n" + "="*50)
print("4. ФИЛЬТРАЦИЯ")

# Простые фильтры
high_return = df[df['daily_return'] > 3.0]
print(f"Акции с доходностью > 3%: {len(high_return)}")

undervalued = df[df['pe_ratio'] < 15]
print(f"Недооцененные акции (P/E < 15): {len(undervalued)}")

# Сложные фильтры
good_dividends = df[(df['dividend_yield'] > 0.02) & (df['daily_return'] > 0)]
print(f"Акции с дивидендами > 2% и положительной доходностью: {len(good_dividends)}")

# 5. ГРУППИРОВКА И АГРЕГАЦИЯ
print("\n" + "="*50)
print("5. ГРУППИРОВКА И АГРЕГАЦИЯ")

# Группировка по компаниям
company_stats = df.groupby('company').agg({
    'close_price': ['mean', 'std', 'min', 'max'],
    'daily_return': 'mean',
    'volume': 'sum',
    'pe_ratio': 'median'
}).round(2)

company_stats.columns = ['avg_price', 'price_std', 'min_price', 'max_price', 'avg_return', 'total_volume', 'median_pe']
print("Статистика по компаниям:")
print(company_stats)

# Группировка по секторам
sector_analysis = df.groupby('sector').agg({
    'daily_return': ['mean', 'std'],
    'pe_ratio': 'mean',
    'dividend_yield': 'mean'
}).round(3)

print("\nАнализ по секторам:")
print(sector_analysis)

# 6. СОРТИРОВКА
print("\n" + "="*50)
print("6. СОРТИРОВКА ДАННЫХ")

# Топ акций по доходности
top_performers = df.nlargest(5, 'daily_return')[['company', 'sector', 'daily_return', 'pe_ratio']]
print("Топ-5 акций по доходности:")
print(top_performers)

# Акции с наименьшим P/E
cheapest = df.nsmallest(5, 'pe_ratio')[['company', 'pe_ratio', 'daily_return']]
print("\nТоп-5 самых дешевых акций (по P/E):")
print(cheapest)

# 7. PIVOT ТАБЛИЦЫ
print("\n" + "="*50)
print("7. PIVOT ТАБЛИЦЫ")

# Сводная таблица: средняя доходность по компаниям и секторам
pivot_return = df.pivot_table(
    values='daily_return', 
    index='company', 
    columns='sector', 
    aggfunc='mean',
    fill_value=0
).round(2)

print("Средняя доходность по компаниям и секторам:")
print(pivot_return)

# 8. ОБРАБОТКА ВРЕМЕННЫХ РЯДОВ
print("\n" + "="*50)
print("8. АНАЛИЗ ВРЕМЕННЫХ РЯДОВ")

df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')

# Добавляем недельные группы
df['week'] = df['date'].dt.isocalendar().week

# Анализ по неделям
weekly_analysis = df.groupby('week').agg({
    'daily_return': 'mean',
    'volume': 'sum',
    'company': 'count'
}).rename(columns={'company': 'trade_count'}).round(2)

print("Анализ по неделям:")
print(weekly_analysis)

# 9. КОНКАТЕНАЦИЯ DATA FRAMES
print("\n" + "="*50)
print("9. КОНКАТЕНАЦИЯ")

# Создаем дополнительный DataFrame с фундаментальными показателями
fundamental_data = pd.DataFrame({
    'company': ['AAPL', 'GOOGL', 'MSFT', 'TSLA', 'AMZN'],
    'market_cap_billion': [2800, 1800, 2200, 800, 1600],
    'employees': [164000, 156500, 221000, 127855, 1541000],
    'founded_year': [1976, 1998, 1975, 2003, 1994]
})

# Объединяем с основными данными
enhanced_df = df.merge(fundamental_data, on='company', how='left')
print("Данные после объединения с фундаментальными показателями:")
print(enhanced_df[['company', 'sector', 'daily_return', 'market_cap_billion']].head())

# 10. СОХРАНЕНИЕ РЕЗУЛЬТАТОВ
print("\n" + "="*50)
print("10. СОХРАНЕНИЕ РЕЗУЛЬТАТОВ")

# Сохраняем различные результаты
enhanced_df.to_csv('pandas_data/enhanced_analysis.csv', index=False)
company_stats.to_csv('pandas_data/company_statistics.csv')
top_performers.to_csv('pandas_data/top_performers.csv', index=False)

print("Результаты сохранены:")
print("- enhanced_analysis.csv - обогащенные данные")
print("- company_statistics.csv - статистика по компаниям") 
print("- top_performers.csv - лучшие акции")

# 11. ПРОДВИНУТАЯ ФИЛЬТРАЦИЯ
print("\n" + "="*50)
print("11. ПРОДВИНУТАЯ ФИЛЬТРАЦИЯ С QUERY")

# Используем метод query для сложных условий
ideal_stocks = df.query(
    'pe_ratio < 20 and daily_return > 0 and dividend_yield > 0.01 and volume > 10000000'
)

print(f"Идеальные акции для инвестирования: {len(ideal_stocks)}")
if len(ideal_stocks) > 0:
    print(ideal_stocks[['company', 'sector', 'daily_return', 'pe_ratio', 'dividend_yield']].round(2))

print(f"\nPandas анализ завершен! Обработано {len(df)} записей.")
```

```python
def rotation_with_lists(matrix):
    n = len(matrix)
    for i in range(int(len(matrix)/2)):
        for j in range(i, len(matrix) - 1-i):
            temp_1 = a[i][n-j-1]
            a[i][n-j-1] = a[i][j]
            temp_2 = a[n-i-1][n-1-j]
            a[n-i-1][n-1-j] = temp_1
            temp_1 = a[n-i-1][j]
            a[n-i-1][j] = temp_2
            a[i][j] = temp_1
    return matrix
```
